```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df = pd.read_csv('twitter_training.csv', header=None)
```

```python
df.head()
```

```python
df.columns = ['id', 'topic', 'sentiment', 'tweet']
```

```python
df.head()
```

```python
df['sentiment'].value_counts()
```

```python
df['sentiment'].value_counts().plot(kind='bar', figsize=(8, 5))
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Number of Tweets')
plt.show()
```

```python
df['tweet'] = df['tweet'].fillna("").astype(str)
all_words = " ".join(df['tweet'])
```

```python
all_words
```

```python
wc = WordCloud(width=1200, height=700, background_color='white').generate(all_words)

plt.figure(figsize=(12, 7))
plt.imshow(wc)
plt.axis('off')
plt.show()
```

```python
positive_text = " ".join(df[df['sentiment']=='Positive']['tweet'])
wc1 = WordCloud(background_color='white').generate(positive_text)
plt.imshow(wc1)
plt.axis('off')
```

```python
negative_text = " ".join(df[df['sentiment']=='Negative']['tweet'])
wc2 = WordCloud(background_color='white').generate(negative_text)
plt.imshow(wc2)
plt.axis('off')
```
